Gerar a tabela básica usando o arquido dados.csv:

```python
import pandas as pd
dados_df = pd.read_csv('dados.csv')
display(dados_df)
```

Valores Mínimos (Math: 0/ Reading: 17/ Writing: 10)
Valores Médios  (Math: 66/Reading: 69/ Writing: 69)
Valores Máximos (Math: 100/Reading:100/Writing: 100)

Usando .describe()

```python
display(dados_df.describe())
```

Notas de Matemática agrupadas por etinía 

```python
math_score_by_ethnie = dados_df[['race/ethnicity', 'math score']].groupby('race/ethnicity')
display(math_score_by_ethnie.describe())
```

Notas de Escrita agrupadas pela Escolaridade dos Pais

```python
parenting_and_writing = dados_df[['parental level of education', 'writing score']].groupby('parental level of education')
display(parenting_and_writing.describe())
```

Notas de Leitura agrupadas por grupos de alunos que tiveram ou não curso preparatório

```python
course_and_reading = dados_df[['test preparation course', 'reading score']].groupby('test preparation course')
display(course_and_reading.describe())
```
